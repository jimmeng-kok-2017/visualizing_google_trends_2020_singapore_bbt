```python
# IMPORT PACKAGES
import pandas as pd
import nltk
```

```python
# READ DATA
bbt_brands_data = pd.read_excel("BBT_Singapore_5_Brands.xlsx")
bbt_brands_data
```

```python
# LOWERCASE CONVERSION
bbt_brands_data['Name 5 Bubble Tea Brands that comes to your mind (Please write in this format: "xxx, yyy, zzz, aaa, bbb". Thank you!)'] = bbt_brands_data['Name 5 Bubble Tea Brands that comes to your mind (Please write in this format: "xxx, yyy, zzz, aaa, bbb". Thank you!)'].apply(lambda x: " ".join(str(x).lower() for x in str(x).split()))
```

```python
# CREATE NEW DATAFRAME TO GET ALL INDIVIDUAL BUBBLE TEA BRAND IN A NEW COLUMN
new_bbt_brands_data = pd.DataFrame(bbt_brands_data['Name 5 Bubble Tea Brands that comes to your mind (Please write in this format: "xxx, yyy, zzz, aaa, bbb". Thank you!)'].str.split(',').tolist(), index=bbt_brands_data.Timestamp).stack()
new_bbt_brands_data = new_bbt_brands_data.reset_index([0, 'Timestamp'])
new_bbt_brands_data.columns = ['Timestamp', 'BBT']
new_bbt_brands_data
```

```python
# CLEAN ALL THE LEFT AND RIGHT SPACES IN THE DATAFRAME
new_bbt_brands_data['BBT'] = new_bbt_brands_data['BBT'].str.strip()
```
